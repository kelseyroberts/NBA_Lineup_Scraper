```python
from selenium import webdriver
from bs4 import BeautifulSoup as soup 
import pandas as pd
from selenium.webdriver.support.ui import Select
import re
```

```python
### connecting to web driver
driver = webdriver.Chrome()
##advanced player stats
driver.get('https://stats.nba.com/lineups/traditional/')
```

```python
### find drop down and select
ddelement= Select(driver.find_element_by_css_selector('body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select'))
##find all and select
All_Element= driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select > option:nth-child(1)").click()
```

```python
##plyr = driver.find_elements_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[1]/div[1]/table/tbody/tr')
table=driver.find_elements_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[2]/div[1]/table/tbody/tr')
links = []
import collections
for telement in table:
    table_dict = collections.OrderedDict()
    Lineups= telement.find_element_by_xpath('.//td[1]').text
    Team=telement.find_element_by_xpath('.//td[2]').text
    GP=telement.find_element_by_xpath('.//td[3]').text
    Min=telement.find_element_by_xpath('.//td[4]').text
    Points=telement.find_element_by_xpath('.//td[5]').text
    FGM=telement.find_element_by_xpath('.//td[6]').text
    FGA=telement.find_element_by_xpath('.//td[7]').text
    FGP=telement.find_element_by_xpath('.//td[8]').text
    F3PM=telement.find_element_by_xpath('.//td[9]').text
    F3PA=telement.find_element_by_xpath('.//td[10]').text
    F3PP=telement.find_element_by_xpath('.//td[11]').text
    FTM=telement.find_element_by_xpath('.//td[12]').text
    FTA=telement.find_element_by_xpath('.//td[13]').text
    FTP=telement.find_element_by_xpath('.//td[14]').text
    OREB=telement.find_element_by_xpath('.//td[15]').text
    DREB=telement.find_element_by_xpath('.//td[16]').text
    REB=telement.find_element_by_xpath('.//td[17]').text
    AST=telement.find_element_by_xpath('.//td[18]').text
    TOV=telement.find_element_by_xpath('.//td[19]').text
    STL=telement.find_element_by_xpath('.//td[20]').text
    BLK=telement.find_element_by_xpath('.//td[21]').text
    BLKA=telement.find_element_by_xpath('.//td[22]').text
    PF=telement.find_element_by_xpath('.//td[23]').text
    PFD=telement.find_element_by_xpath('.//td[24]').text
    P_M=telement.find_element_by_xpath('.//td[25]').text
    links.append([Lineups,Team,GP,Min,Points,FGM,FGA,FGP,F3PM,F3PA,F3PP,FTM,FTA,
                 FTP,OREB,DREB,REB,AST,TOV,STL,BLK,BLKA,PF,
                 PFD,P_M])
```

```python
Lineup_Stats_Trad=pd.DataFrame(links)
```
