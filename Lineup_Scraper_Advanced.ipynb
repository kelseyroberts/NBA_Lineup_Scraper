```python
from selenium import webdriver
from bs4 import BeautifulSoup as soup 
import pandas as pd
from selenium.webdriver.support.ui import Select
import re
```

```python
### connecting to web driver
driver = webdriver.Chrome()
##advanced player stats
driver.get('https://stats.nba.com/lineups/advanced/?sort=TEAM_ABBREVIATION&dir=-1')
```

```python
### find drop down and select
ddelement= Select(driver.find_element_by_css_selector('body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select'))
##find all and select
All_Element= driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select > option:nth-child(1)").click()
```

```python
##plyr = driver.find_elements_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[1]/div[1]/table/tbody/tr')
table=driver.find_elements_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[2]/div[1]/table/tbody/tr')
links = []
import collections
for telement in table:
    table_dict = collections.OrderedDict()
    Lineups= telement.find_element_by_xpath('.//td[1]').text
    Team=telement.find_element_by_xpath('.//td[2]').text
    GP=telement.find_element_by_xpath('.//td[3]').text
    Min=telement.find_element_by_xpath('.//td[4]').text
    OFFRTG=telement.find_element_by_xpath('.//td[5]').text
    DEFRTG=telement.find_element_by_xpath('.//td[6]').text
    NETRTG=telement.find_element_by_xpath('.//td[7]').text
    AST_Per=telement.find_element_by_xpath('.//td[8]').text
    AST_TO=telement.find_element_by_xpath('.//td[9]').text
    AST_Ratio=telement.find_element_by_xpath('.//td[10]').text
    OREB_Per=telement.find_element_by_xpath('.//td[11]').text
    DREB_Per=telement.find_element_by_xpath('.//td[12]').text
    REB_Per=telement.find_element_by_xpath('.//td[13]').text
    TO_Ratio=telement.find_element_by_xpath('.//td[14]').text
    EFG_Per=telement.find_element_by_xpath('.//td[15]').text
    TS_Per=telement.find_element_by_xpath('.//td[16]').text
    Pace=telement.find_element_by_xpath('.//td[17]').text
    PIE=telement.find_element_by_xpath('.//td[18]').text
    links.append([Lineups,Team,GP,Min,OFFRTG,DEFRTG,NETRTG,AST_Per,AST_TO,AST_Ratio,OREB_Per,
                  DREB_Per,REB_Per,TO_Ratio,EFG_Per,TS_Per,Pace,PIE])
```

```python
Lineup_Stats=pd.DataFrame(links)
```
