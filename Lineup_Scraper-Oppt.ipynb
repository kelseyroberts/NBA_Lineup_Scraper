```python
from selenium import webdriver
from bs4 import BeautifulSoup as soup 
import pandas as pd
from selenium.webdriver.support.ui import Select
import re
```

```python
### connecting to web driver
driver = webdriver.Chrome()
##advanced player stats
driver.get('https://stats.nba.com/lineups/opponent/')
```

```python
### find drop down and select
ddelement= Select(driver.find_element_by_css_selector('body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select'))
##find all and select
All_Element= driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select > option:nth-child(1)").click()
```

```python
##plyr = driver.find_elements_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[1]/div[1]/table/tbody/tr')
table=driver.find_elements_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[2]/div[1]/table/tbody/tr')
links = []
import collections
for telement in table:
    table_dict = collections.OrderedDict()
    Lineups= telement.find_element_by_xpath('.//td[1]').text
    Team=telement.find_element_by_xpath('.//td[2]').text
    GP=telement.find_element_by_xpath('.//td[3]').text
    Min=telement.find_element_by_xpath('.//td[4]').text
    OPP_FGM=telement.find_element_by_xpath('.//td[5]').text
    OPP_FGA=telement.find_element_by_xpath('.//td[6]').text
    OPP_FGP=telement.find_element_by_xpath('.//td[7]').text
    OPP_3PM=telement.find_element_by_xpath('.//td[8]').text
    OPP_3PA=telement.find_element_by_xpath('.//td[9]').text
    OPP_3PP=telement.find_element_by_xpath('.//td[10]').text
    OPP_FTM=telement.find_element_by_xpath('.//td[11]').text
    OPP_FTA=telement.find_element_by_xpath('.//td[12]').text
    OPP_FTP=telement.find_element_by_xpath('.//td[13]').text
    OPP_OREB=telement.find_element_by_xpath('.//td[14]').text
    OPP_DREB=telement.find_element_by_xpath('.//td[15]').text
    OPP_REB=telement.find_element_by_xpath('.//td[16]').text
    OPP_AST=telement.find_element_by_xpath('.//td[17]').text
    OPP_TOV=telement.find_element_by_xpath('.//td[18]').text
    OPP_STL=telement.find_element_by_xpath('.//td[19]').text
    OPP_BLK=telement.find_element_by_xpath('.//td[20]').text
    OPP_BLKA=telement.find_element_by_xpath('.//td[21]').text
    OPP_PF=telement.find_element_by_xpath('.//td[22]').text
    OPP_PFD=telement.find_element_by_xpath('.//td[23]').text
    OPP_PTS=telement.find_element_by_xpath('.//td[24]').text
    OPP_P_M=telement.find_element_by_xpath('.//td[25]').text
    links.append([Lineups,Team,GP,Min,OPP_FGM,OPP_FGA,OPP_FGP,OPP_3PM,OPP_3PA,OPP_3PP,OPP_FTM,OPP_FTA,
                 OPP_FTP,OPP_OREB,OPP_DREB,OPP_REB,OPP_AST,OPP_TOV,OPP_STL,OPP_BLK,OPP_BLKA,OPP_PF,
                 OPP_PFD,OPP_PTS,OPP_P_M])
```

```python
Lineup_Stats_Oppt=pd.DataFrame(links)
```
